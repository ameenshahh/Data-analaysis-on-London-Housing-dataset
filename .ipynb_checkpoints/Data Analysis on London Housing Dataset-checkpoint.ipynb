# Data Analysis on London Housing Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("5. London Housing Data.csv")
```

```python
data
```

```python
#null values in each column
data.isnull().sum()
```

```python
sns.heatmap(data.isnull())
plt.show()
```

```python
# 1. convert the datatype of date column to date-Time format

data.head()
```

```python
data.dtypes
```

```python
data.date = pd.to_datetime(data.date)
```

```python
# 2.1 Adding a new column in the dataframe which contains year only
data['year'] = data.date.dt.year
data
```

```python
# 2.2 Add a new column "month" as 2nd column in the dataframe,which contains month only
data.insert(1, 'month', data.date.dt.month)
```

```python
data
```

```python
# 3. remove the columns 'year' and 'month' from the dataframe
data.drop(['month','year'], axis = 1,inplace = True)
```

```python
data
```

```python
# 4. show all the the records where no of crimes is 0 
#    and how many such records are there

data[data['no_of_crimes'] == 0]
```

```python
# 5 what is the maximum and minimum average price per year in england
data['year'] = data.date.dt.year
df1 = data[data.area == 'england']
df1.groupby('year').average_price.max()
```

```python
# 6. what is the maximum and minimum no of crimes recorded per year
data.groupby('area').no_of_crimes.max().sort_values(ascending = True)
```

```python
# 7. show the total count of records of each area, where average price is less than 100000
data[data.average_price < 100000].area.value_counts()
```
